```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv"
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,Y,train_size=0.7,shuffle=True)
```

```python
LinearReg = LinearRegression()

LinearReg.fit(X_train,Y_train)
```

```python
Y_Pred = LinearReg.predict(X_test)
```

```python
LinearRegression_MAE = metrics.mean_absolute_error(Y_test,Y_Pred)
LinearRegression_MSE = metrics.mean_squared_error(Y_test,Y_Pred)
LinearRegression_R2 = LinearReg.score(X_test,Y_test)
```

```python
LinearRegression_MAE
```

```python
LinearRegression_MSE
```

```python
LinearRegression_R2
```

```python
Knn=KNeighborsClassifier(n_neighbors=4)

Knn.fit(X_train,Y_train)
```

```python
Y_predKnn = Knn.predict(X_test)
```

```python
KNN_Accuracy_Score = accuracy_score(Y_test,Y_predKnn)
KNN_JaccardIndex = jaccard_score(Y_test,Y_predKnn)
KNN_F1_Score = f1_score(Y_test,Y_predKnn)
```

```python
KNN_Accuracy_Score
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,Y,test_size=0.2,random_state=1)
```

```python
LR=LogisticRegression(solver='liblinear')
LR.fit(X_train,Y_train)
```

```python
predictions = LR.predict(X_test)

predictions_Prob = LR.predict_proba(X_test)
```

```python
LR_Accuracy_Score = accuracy_score(Y_test,predictions)
LR_JaccardIndex = jaccard_score(Y_test,predictions)
LR_F1_Score = f1_score(Y_test,predictions)
LR_Log_Loss = log_loss(Y_test,predictions_Prob)
```

```python
SVM = svm.LinearSVC()
SVM.fit(X_train,Y_train)
```

```python
PredictionSVM= SVM.predict(X_test)
```

```python
SVM_Accuracy_Score = accuracy_score(Y_test,PredictionSVM)
SVM_JaccardIndex = jaccard_score(Y_test,PredictionSVM)
SVM_F1_Score = f1_score(Y_test,PredictionSVM)
```

```python
tree= DecisionTreeClassifier()

tree.fit(X_train,Y_train)
```

```python
predictionTree=tree.predict(X_test)
```

```python
Tree_Accuracy_Score = accuracy_score(Y_test,predictionTree)
Tree_JaccardIndex = jaccard_score(Y_test,predictionTree)
Tree_F1_Score = f1_score(Y_test,predictionTree)
```
